```python
from skimage import data, color
from skimage.transform import rescale, resize, downscale_local_mean
```

```python
#dir(data)
p = "{n:<15s} Image type: {t:<15s} Image shape:{s:<10s}".format 
d = [data.brick, data.astronaut, data.binary_blobs, data.cell, 
     data.checkerboard, data.chelsea, data.camera, 
    data.clock, data.coffee, data.coins, data.colorwheel,
    data.horse, data.hubble_deep_field, data.immunohistochemistry, 
     data.logo, data.moon, data.microaneurysms]

n = len(d)
nc=5
nr = n//nc if n%nc==0 else  (n//nc)+1
f, axs = plt.subplots(nr, nc, figsize=(nc*3, nr*3), )
rs = 0.25
for i, f in enumerate(d):
    img = f()
    nm = f"{str(f).split(' ')[1]}"
    print( f"{i+1}.", p(n=nm, t=f"{type(img)}", s=f"{img.shape}" ))
    ax = axs[ i//nc, i%nc]
#     ax.imshow( rescale( color.rgb2gray(img), rs, anti_aliasing=False ) )
    ax.imshow( img,  aspect='auto')
    ax.set_title( nm ) 
    ax.axis('off')

plt.show(); 
```
